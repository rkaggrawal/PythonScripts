## Single Linked List

```python
class Node():
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next
        
class LinkedList():
    def __init__(self):
        self.head = Node()
        
    def display(self):
        elems=[]
        curr_node = self.head
#         print("Data is {}".format(curr.data))
#         print("Data is {}".format(curr.next))
        while(curr_node.next != None):
            curr_node = curr_node.next
            elems.append(curr_node.data)
        print(elems)
    
    def insert_last(self, data):
        curr_node = self.head
        while(curr_node.next != None):
            curr_node = curr_node.next
        new_node = Node(data)
        curr_node.next = new_node
    
    def insert_first(self, data):
        curr_node = self.head
        new_node = Node(data, curr_node.next)
        curr_node.next = new_node
    
    def insert_before(self, data, item):
        curr_node = self.head
        while curr_node.next !=None:
            last_node = curr_node
            curr_node = curr_node.next
            if curr_node.data == item:
                new_node = Node(data, last_node.next)
                last_node.next = new_node

    def insert_after(self, data, item):
        curr_node = self.head
        while curr_node.next !=None:
            if curr_node.data == item:
                new_node = Node(data, curr_node.next)
                curr_node.next = new_node
                break
            curr_node = curr_node.next


    def remove(self, item):
        curr_node = self.head
        while curr_node.next !=None:
            last_node = curr_node
            curr_node = curr_node.next
            if curr_node.data == item:
                last_node.next = curr_node.next
                return
 
    def search(self, item):
        curr_node = self.head
        while curr_node.next != None:            
            if curr_node.data == item:
                print("Node found {}".format(item))
                break
            curr_node = curr_node.next
        else:
            if curr_node.data == item:
                print("Node found {}".format(item))
            else:
                print("Node not found {}".format(item))

if __name__ == '__main__':
    llobj = LinkedList()
#     llobj.insert_last(5)
#     llobj.insert_last(7)
#     llobj.insert_first(10)
#     llobj.insert_before(12,7)
#     llobj.insert_after(15,10)
#     llobj.insert_after(20,12)
#     llobj.remove(5)
    llobj.display()
#     llobj.search(40)
```

## Circular Single Linked List

```python
class Node:
    def __init__(self, data=None, next=None):
            self.data = data
            self.next = next
            
class CirLinkedList:
    def __init__(self):
        self.head = Node()
        self.tail = Node()
        self.head.next = self.tail
        self.tail.next = self.head
    
    def insert_last(self, data):
        newNode = Node(data);
        if self.head.data is None:    
            self.head = newNode;    
            self.tail = newNode;    
            newNode.next = self.head;
        else:
            self.tail.next = newNode;
            self.tail = newNode;
            self.tail.next = self.head;
    
    def display(self):
        elems = []
        curr_node = self.head
        if self.head is None:
            print("List is empty");    
            return;    
        else:
#             print(curr_node.data),
            while(curr_node.next != self.head):    
                elems.append(curr_node.data)
                curr_node = curr_node.next;  
            else:
                elems.append(curr_node.data)
                print(elems)

        
if __name__ == '__main__':
    cllobj = CirLinkedList()
    cllobj.insert_last(10)
    cllobj.insert_last(20)
    cllobj.insert_last(30)
    cllobj.display()
```

## Doubly Linked List

```python
class Node():
    def __init__(self, data=None, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev
        
class DoublyLinkedList():
    def __init__(self):
        self.head = Node()
        
    def display(self):
        elems=[]
        curr_node = self.head
#         print("Data is {}".format(curr.data))
#         print("Data is {}".format(curr.next))
        while(curr_node.next != None):
            last_node = curr_node
            curr_node = curr_node.next
            curr_node.prev = last_node
            elems.append(curr_node.data)
        print(elems)
        
    def insert_last(self, data):
        curr_node = self.head
        while(curr_node.next != None):
            last_node = curr_node
            curr_node = curr_node.next
            curr_node.prev = last_node
        new_node = Node(data)
        curr_node.next = new_node
        new_node.prev = curr_node
    
    def insert_first(self, data):
        curr_node = self.head
        new_node = Node(data, curr_node.next)
        new_node.next = curr_node.next
        curr_node.next = new_node
    
    def insert_before(self, data, item):
        curr_node = self.head
        while curr_node.next !=None:
            last_node = curr_node
            curr_node = curr_node.next
            if curr_node.data == item:
                new_node = Node(data, last_node.next, last_node)
                last_node.next = new_node
    
    def insert_after(self, data, item):
        curr_node = self.head
        while curr_node.next !=None:
            if curr_node.data == item:
                new_node = Node(data, curr_node.next, curr_node)
                curr_node.next = new_node
                break
            curr_node = curr_node.next
    
    def remove(self, item):
        curr_node = self.head
        while curr_node.next !=None:
            last_node = curr_node
            curr_node = curr_node.next
            next_node = curr_node.next
            if curr_node.data == item:
                last_node.next = curr_node.next
                next_node.prev = last_node

    def search(self, item):
        curr_node = self.head
        while curr_node.next != None:            
            if curr_node.data == item:
                print("Node found {}".format(item))
                break
            curr_node = curr_node.next
        else:
            if curr_node.data == item:
                print("Node found {}".format(item))
            else:
                print("Node not found {}".format(item))
                
                
if __name__ == '__main__':
    dllobj = DoublyLinkedList()
#     dllobj.insert_last(5)
#     dllobj.insert_last(8)
#     dllobj.insert_first(9)
#     dllobj.insert_before(18, 8)
#     dllobj.insert_after(20, 5)
#     dllobj.remove(20)
    dllobj.display()
#     dllobj.search(18)
#     dllobj.search(20)
```

## Binary Search Tree - way1

```python
class Node():
    def __init__(self, value=None):
        self.value = value
        self.left_child = None
        self.right_child = None
        
    def insert(self, data):
        if self.value == data:
            return False
        elif data < self.value:
            if self.left_child == None:
                self.left_child = Node(data)
                return True
            else:
                return self.left_child.insert(data)
        else:
            if self.right_child == None:
                self.right_child = Node(data)
                return True
            else:
                return self.right_child.insert(data)
            
    def find(self, data):
        if(self.value == data):
            return True
        elif data < self.value:
            if self.left_child:
                return self.left_child.find(data)
            else:
                return False
        else:
            if self.right_child:
                return self.right_child.find(data)
            else:
                return False
            
    def inorder(self):
        if self:
            if self.left_child:
                self.left_child.inorder()
            print(self.value, end=' ')
            if self.right_child:
                self.right_child.inorder()
            
class BST():
    def __init__(self):
        self.root = None
    
    def insert(self, data):
        if self.root:
            return self.root.insert(data)
        else:
            self.root = Node(data)
            return True
        
    def find(self, data):
        if self.root.find(data):
            print("\nFound {}".format(data))
        else:
            print("Not Found {}".format(data))
        
    def inorder(self):
        print("preorder")
#         print("self.root.value %d" %(self.root.value))
        self.root.inorder()
        
bst_tree = BST()
list = [10,8,12,6,3,21,14,99,18,7,6]
for x in list:
    bst_tree.insert(x)

# Following both statements for calling class method are same.
# bst_tree.inorder() # method1  - obj.method()
BST.inorder(bst_tree) #method2 - class.method(obj)
bst_tree.find(14)
bst_tree.find(50)
```

## Binary Search Tree - way2

```python
class Node():
    def __init__(self, value=None):
        self.value = value
        self.left_child = None
        self.right_child = None
        
class BST():
    def __init__(self):
        self.root = None
    
    def insert(self,value):
        if self.root == None:
            self.root = Node(value)
        else:
            self._insert(value, self.root)
    
    def _insert(self, value, curr_node):
        if value < curr_node.value:
            if curr_node.left_child == None:
                curr_node.left_child = Node(value)
            else:
                self._insert(value, curr_node.left_child)
        
        elif value > curr_node.value:
            if curr_node.right_child == None:
                curr_node.right_child = Node(value)
            else:
                self._insert(value, curr_node.right_child)   
        else:
            print("value {} already exists".format(value))
    
    def print_tree(self):
        if self.root != None:
            self._print_tree(self.root)
        else:
            print("Tree is empty")
    
    def _print_tree(self, curr_node):
        if curr_node != None:
            self._print_tree(curr_node.left_child)
            print(curr_node.value)
            self._print_tree(curr_node.right_child)
            
            
tree = BST()
list = [10,8,12,6,3,21,14,99,18,7,6]

for x in list:
    tree.insert(x)
tree.print_tree()

    
```

##  Binary Search Tree - way3

```python
class BST:
    def __init__(self, data=None):
        self.left = None
        self.right = None
        self.data = data

    def insert(self, data):
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = BST(data)
                else:
                    self.left.insert(data)
            elif data > self.data:
                if self.right is None:
                    self.right = BST(data)
                else:
                    self.right.insert(data)

            else:
                print ("Value already exists")

        else:
            self.data = data

    def print_tree(self):
        if self.left:
            self.left.print_tree()
        print (self.data)
        if self.right:
            self.right.print_tree()

tree = BST()
list = [10,8,12,6,3,21,14,99,18,7,6]

for x in list:
    tree.insert(x)
tree.print_tree()
```
